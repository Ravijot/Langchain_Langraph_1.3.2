## STORE

Stores let agents persist information across threads, including user preferences, accumulated knowledge, and   
facts that should survive beyond a single conversation. Unlike checkpointers, which save the full graph state   
 scoped to one thread, stores hold arbitrary key-value data accessible from any thread.

```python
from langgraph.store.memory import InMemoryStore
store = InMemoryStore()
```

##  NAMESPACE 

Each memory is organized under a custom namespace (similar to a folder) and a distinct key (like a file name).  
Namespaces often include user or org IDs or other labels that makes it easier to organize information.  


```python
user_id = "1"
namespace_for_memory = (user_id, "memories")
```

Use the store.put method to save memories to the namespace in the store.  
Specify the namespace, as defined above, and a key-value pair for the memory: the key is simply a unique identifier for the memory    
(memory_id) and the value (a dictionary) is the memory itself.

```python
import uuid
memory_id = str(uuid.uuid4())
memory = {"food_preference" : "I like pizza"}
store.put(namespace_for_memory, memory_id, memory)
```

```python
memories = store.search(namespace_for_memory)
```

```python
print(memories)
```

```python
print(store.search(namespace_for_memory, limit=2))
```
